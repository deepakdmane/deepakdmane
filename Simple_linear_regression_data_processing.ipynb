# import Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
import csv
import random

# Generate 300 records
num_records = 300
data = [("Experience", "Salary")]

for _ in range(num_records):
    # Generate experience: between 0.0 and 10.0 years, rounded to 1 decimal
    experience = round(random.uniform(0, 10), 1)

    # Generate salary based on experience range
    base_salary = 250000
    salary = int(base_salary + experience * random.randint(70000, 120000))

    data.append((experience, salary))

# Save to CSV
file_path = "salary_data.csv"
with open(file_path, "w", newline="") as file:
    writer = csv.writer(file)
    writer.writerows(data)

file_path
```

# Import DataSET

```python
try:
    dataset = pd.read_csv('salary_data.csv')
    display(dataset.head())
except FileNotFoundError:
    print("Error: 'Data.csv' not found. Please ensure the file exists in the current directory.")
except pd.errors.EmptyDataError:
    print("Error: 'Data.csv' is empty.")
except pd.errors.ParserError:
    print("Error: 'Data.csv' could not be parsed. Please check the file format.")
except Exception as e:
    print(f"An unexpected error occurred: {e}")

x = dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1].values
```

```python
print(x)
```

```python
print(y)
```

# Splitting data into training data and test data

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 0)
```

```python
X_train, X_test, y_train, y_test
```

# Feature Scaling

```python
from sklearn.linear_model import LinearRegression
regression = LinearRegression()
regression.fit(X_train, y_train)
```

```python
ypred = regression.predict(X_test)
```

```python
plt.scatter(X_train, y_train, color = 'red')
plt.plot(X_train, regression.predict(X_train), color = 'blue')
plt.title('Salary vs Experience (Training set)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show()
```

```python
plt.scatter(X_test, y_test, color = 'red')
plt.plot(X_train, regression.predict(X_train), color = 'blue')
plt.title('Salary vs Experience (Training set)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show()
```

```python
prediction = regression.predict([[10]])  # Input must be 2D (e.g., [[12]])
print(prediction)  # Output: [90000]
```

```python
# Assuming your model is named "model"
b0 = regression.intercept_   # Intercept (b₀)
b1 = regression.coef_[0]     # Slope (b₁) for the first feature (e.g., "Years of Experience")

print(f"Final equation: Salary = {b0:.2f} + {b1:.2f} × Experience")
# Example output: Salary = 30000.00 + 5000.00 × Experience
```

```python
import statsmodels.api as sm

# Add a constant (intercept term) to the features
X = sm.add_constant(x)  # X is your feature matrix (e.g., years of experience)
model = sm.OLS(y, X).fit()  # y is the target (e.g., salary), Use X here instead of x

# Get coefficients
b0 = model.params[0]  # Intercept
b1 = model.params[1]  # Slope

print(model.summary())  # Displays full regression results (including b₀ and b₁)
```
